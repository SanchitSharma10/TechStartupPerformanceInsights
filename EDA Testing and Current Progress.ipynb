```python
for industry in industries:
    # select the rows for the current industry
    industry_df = df[df['Industry'] == industry]
    # select the starting point at random
    start = np.random.randint(0, interval)
    # select the samples using systematic sampling
    samples.extend(industry_df.iloc[start::interval, :])
print(samples)
```

```python
# create a stratified sample
samples = df.groupby('industry').apply(lambda x: x.sample(n=1))
```

```python
# create a stratified sample by grouping the data in the Dataframe(df) by the "Industry" column. 
# Then, for each group it will apply a function using the apply method, and sample method
# the lambda in this case is for defining anonymous functions (kind of like a throw-away)
# with this the sample dataframe will now have one row for each unique value in the Industry column.
samples = df.groupby('Industry').apply(lambda x: x.sample(n=1))
```
